# Building a linear regression ML model to predict the rental prices of property in Manhattan

The data used has come from StreetEasy: https://streeteasy.com/blog/download-data/

In this project, I set out to create a linear regression machine learning model that can predict the monthly rental value of Manhattan properties, depending on their features.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

df = pd.read_csv('manhattan.csv')

df.columns
```

Setting the variables that will be used to predict the rental price

```python
x = df[['bedrooms', 'bathrooms', 'size_sqft','min_to_subway', 'floor', 'building_age_yrs', 'no_fee', 'has_roofdeck', 'has_washer_dryer', 'has_doorman', 'has_elevator', 'has_dishwasher', 'has_patio', 'has_gym']]

y = df['rent']
```

Splitting the data into a training set and a test set so that the model accuracy can be validated 

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2)
```

Creating the LR model and fitting the data to it

```python
mlr = LinearRegression()
mlr.fit(x_train, y_train)
```

By plotting the predicted rental results against the actual rental results we can already see model showing strong correlation to the actual data

```python
y_predict = mlr.predict(x_test)

plt.scatter(y_test,y_predict, alpha=0.6)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual VS Predicted Rent')
plt.show()
```

Next, I looked at the models coefficient values to see if any features don't behave linearly. It can help to increase the accuracy of the model by finding and removing them from the features list.

```python
mlr.coef_
```

Scores for the training and testing data

```python
print(mlr.score(x_test, y_test))
print(mlr.score(x_train, y_train))
```

By looking at the coefficient weighting, features 3,6,11 and 12 might be worth investigating

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
plt.scatter(df[['size_sqft']], df[['rent']], alpha=0.4)
plt.subplot(2,2,2)
plt.scatter(df[['building_age_yrs']], df[['rent']], alpha=0.4)
plt.subplot(2,2,3)
plt.scatter(df[['has_elevator']], df[['rent']], alpha=0.1)
plt.subplot(2,2,4)
plt.scatter(df[['has_dishwasher']], df[['rent']], alpha=0.1)
```

After trying and changing the features used there was little improvement in the accuracy of the model

```python
## rebuild model without uncorrelated features washer dryer and dishwasher

x_new = df[['bedrooms', 'bathrooms', 'size_sqft','min_to_subway', 'floor', 'building_age_yrs', 'no_fee', 'has_roofdeck', 'has_doorman', 'has_elevator', 'has_patio', 'has_gym']]

x_train1, x_test1, y_train1, y_test1 = train_test_split(x_new, y, train_size=0.8, test_size=0.2)

mlr1 = LinearRegression()
mlr1.fit(x_train1, y_train1)

y_predict1 = mlr1.predict(x_test1)

print(mlr1.score(x_test1, y_test1))
print(mlr1.score(x_train1, y_train1))
mlr1.coef_
```
